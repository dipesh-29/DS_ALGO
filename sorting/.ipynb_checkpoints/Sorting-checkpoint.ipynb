```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr
```

```python
def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i+1, len(arr)):
            if arr[min_index] > arr[j]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

```python
def insertion_sort(arr):
    for i in range(1,len(arr)):
        curr_num = arr[i]
        j = i-1
        
        while curr_num < arr[j] and j >= 0 :
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = curr_num  
    return arr
                
```

```python
def partition(arr,left,right):
    pivot = arr[right]
    j = left
    for i in range(left,right):
        if pivot >= arr[i]:
            arr[i],arr[j] = arr[j],arr[i]
            j += 1
    arr[j],arr[right] = arr[right],arr[j]
    return j

def quick_sort(arr,left,right):
    if left < right : 
        p = partition(arr,left,right)
        quick_sort(arr,left,p-1)
        quick_sort(arr,p+1,right)
        return arr
    
```

```python
def heapify(arr,index,size):
    if arr:
        largest = index
        left = 2*index + 1
        right = 2*index + 2
        
        if left < size and arr[largest] < arr[left] :
            largest = left
        if right < size and arr[largest] < arr[right] :
            largest = right
        if largest != index:
            arr[largest],arr[index] = arr[index], arr[largest]
            heapify(arr,largest,size)
        

def heap_sort(arr):
    size = len(arr)-1
    for index in range(size,-1,-1):
        heapify(arr,index,size)
    
    for i in range(size,0,-1):
        arr[i],arr[0]=arr[0],arr[i]
        heapify(arr,0,i)

    return arr
```

```python
class Sorting:

    def bubble_sort(self, arr):
        for i in range(len(arr)):
            for j in range(i+1, len(arr)):
                if arr[i] > arr[j]:
                    #Swap a[i] and a[j]
                    arr[i], arr[j] = arr[j], arr[i]
        return arr

    def selection_sort(self, arr):
        for i in range(len(arr)):
            min_index = i
            for j in range(i+1, len(arr)):
                if arr[min_index] > arr[j]:
                    min_index = j
            # Swap a[i] and a[min_index]
            arr[i], arr[min_index] = arr[min_index], arr[i]
        return arr

    def insertion_sort(self, arr):
        for i in range(1, len(arr)):
            curr_num = arr[i]
            j = i-1
            while curr_num < arr[j] and j >= 0 :
                arr[j+1] = arr[j]
                j -= 1
            arr[j+1] = curr_num
        return arr

    def partition(self, arr, left, right):
        pivot = arr[right]
        j = left
        for i in range(left, right):
            if pivot >= arr[i]:
                arr[i], arr[j] = arr[j], arr[i]
                j +=1 
        arr[j], arr[right] = arr[right], arr[j]
        return j

    def quick_sort(self, arr, left, right):
        if left < right:
            p = self.partition(arr, left, right)
            self.quick_sort(arr, left, p-1)
            self.quick_sort(arr, p+1, right)
            return arr

    def heapify(self, arr, size, index):
        largest = index
        left = 2*index+1
        right = 2*index+2
        if left <= size and arr[largest] < arr[left]:
            largest = left
        if right <= size and arr[largest] < arr[right]:
            largest = right
        if largest != index:
            arr[index], arr[largest] = arr[largest], arr[index]
            self.heapify(arr, size, largest)
        return arr
    
    def heap_sort(self, arr):
        # Build heap
        size = len(arr)-1
        for index in range(size, -1, -1):
            arr = self.heapify(arr, size, index)

        # Delete each element and call heapify function.
        for i in range(size,0,-1):
            arr[i],arr[0]=arr[0],arr[i]
            self.heapify(arr,i-1,0)
        return arr

    def merge(self, arr, left, mid, right):
        i = left
        j = mid+1
        k = i
        result = [x for x in arr]
        while (i <= mid and j <= right):
            if arr[i] < arr[j]:
                result[k] = arr[i]
                i += 1
            else:
                result[k] = arr[j]
                j += 1
            k += 1
        if i > mid : 
            while j <= right:
                result[k] = arr[j]
                j += 1
                k += 1
        else:
            while i <= mid:
                result[k] = arr[i]
                i += 1
                k += 1
                
        for k in range(left, right+1):
            arr[k] = result[k]
        return arr
        
    def merge_sort(self, arr, left, right):
        if left < right:
            mid = int((left + right) / 2)
            self.merge_sort(arr, left, mid)
            self.merge_sort(arr, mid+1, right)
            self.merge(arr, left, mid, right)
        return arr
        
            

if __name__=="__main__":
    arr = [4,2,5,7,3,2,4,1,8,4,9,6]
    sorting_obj = Sorting()
    #result = sorting_obj.insertion_sort(arr)
    #result = sorting_obj.quick_sort(arr, 0, len(arr)-1)
    result = sorting_obj.merge_sort(arr, 0, len(arr)-1)
    print(result)
```
