```python
# Inorder Traversal :
class TreeTraversal:
    def inorder(self, root, result):
        if root is None:
            return None
        self.inorder(root.left, result)
        result.append(root.val)
        self.inorder(root.right, result)
        return result

    # Given root of tree return list containing its inoder traversal.
    def inorder_call(self, root):
        result = []
        result = self.inorder(root, result)
        return result

    
    def preorder(self, root, result):
        if root is None:
            return None
        result.append(root.val)
        self.preorder(root.left, result)
        self.preorder(root.right, result)
        return result
    
    
    def preorder_call(self, root):
        result = []
        result = self.preorder(root, result)
        return result


    def postorder(self, root, result):
        if root is None:
            return None
        self.postorder(root.left, result)
        self.postorder(root.right, result)
        result.append(root.val)
        return result
    
    
    def postorder_call(self, root):
        result = []
        result = self.postorder(root, result)
        return result


    def levelorder_traversal(self, root):
        if root is None:
            return None
        queue = []
        level_order = []
        queue.append(root)
        while queue:
            root = queue.pop(0)
            level_order.append(root.val)
            if root.left:
                queue.append(root.left)
            if root.right:
                queue.append(root.right)
        return level_order
            
        


tt_obj = TreeTraversal()
result = tt_obj.inorder_call(root)
print(result)
result = tt_obj.preorder_call(root)
print(result)
result = tt_obj.postorder_call(root)
print(result)
result = tt_obj.levelorder_traversal(root)
print(result)
```

# Find Maximum element in binary Tree : 

```python
class TreeQuestions:

    def find_max_element(self, root):
        if root is None:
            return float('-inf')
    
        if not root.left and not root.right :
            return root.val
    
        left_max = self.find_max_element(root.left)
        right_max = self.find_max_element(root.right)
        return max(left_max, right_max)

    def search_element(self, root, element):
        if root is None:
            return None
        if root.val == element:
            return True
        left = self.search_element(root.left, element)
        right = self.search_element(root.right, element)
        if left or right:
            return True
        else:
            return False

    def sizeoftree(self, root):
        if root is None:
            return 0
        return self.sizeoftree(root.left) + self.sizeoftree(root.right) + 1

    def level_order_reverse_order(self, root):
        if root is None:
            return None
        queue = []
        queue.append(root)
        rev_level_order = []
        while queue:
            root = queue.pop(0)
            rev_level_order.append(root.val)
            if root.right:
                queue.append(root.right)
            if root.left:
                queue.append(root.left)
        return rev_level_order

    def height_of_binary_tree(self, root):
        if root is None:
            return 0
        left = self.height_of_binary_tree(root.left)
        right = self.height_of_binary_tree(root.right)
        return max(left, right) + 1

    def deepest_node_binary_tree(self, root):
        if root is None:
            return 0
        queue = []
        queue.append(root)
        queue.append('$')
        deepest_node = None
        while queue:
            root = queue.pop(0)
            if root == '$': 
                if queue:
                    queue.append('$')
            else:
                deepest_node = root.val
                if root.left:
                    queue.append(root.left)
                if root.right:
                    queue.append(root.right)
        return deepest_node

    def deepest_node_binary_tree_optimised(self, root):
        if root is None:
            return 0
        queue = []
        queue.append(root)
        while queue:
            root = queue.pop(0)
            if root.left:
                queue.append(root.left)
            if root.right:
                queue.append(root.right)
        return root.val

    def number_of_leaves_recursion(self, root):
        if root is None:
            return 0
        if root.left is None and root.right is None:
            return 1
        left = self.number_of_leaves_recursion(root.left)
        right = self.number_of_leaves_recursion(root.right)
        return left + right
        

    def number_of_leaves_iterative(self, root):
        if root is None:
            return 0
        queue = []
        queue.append(root)
        num_of_leaves_nodes = 0
        while queue:
            root = queue.pop(0)
            if root.left is None and root.right is None:
                num_of_leaves_nodes += 1
            if root.left:
                queue.append(root.left)
            if root.right:
                queue.append(root.right)
        return num_of_leaves_nodes

    def number_of_full_nodes_recursion(self, root):
        if root is None:
            return 0
        left_leave_nodes = self.number_of_full_nodes_recursion(root.left)
        right_leave_nodes = self.number_of_full_nodes_recursion(root.right)
        if root.left and root.right:
            return 1 + (left_leave_nodes + right_leave_nodes)
        return 0


    def number_of_full_nodes_iterative(self, root):
        if root is None:
            return 0
        queue = []
        queue.append(root)
        num_of_full_nodes = 0
        while queue:
            root = queue.pop(0)
            if root.left and root.right:
                num_of_full_nodes += 1
            if root.left:
                queue.append(root.left)
            if root.right:
                queue.append(root.right)
        return num_of_full_nodes

    # NOT WORKING
    def number_of_half_nodes_recursive(self, root):
        if root is None:
            return 0
        left_leave_nodes = self.number_of_full_nodes_recursion(root.left)
        right_leave_nodes = self.number_of_full_nodes_recursion(root.right)
        if root.left and root.right:
            return (left_leave_nodes + right_leave_nodes)
        if root.left or root.right:
            return 1 + (left_leave_nodes + right_leave_nodes)
        return 0

    def number_of_half_nodes_iterative(self, root):
        if root is None:
            return o
        queue = []
        queue.append(root)
        half_nodes = 0
        while queue:
            root = queue.pop(0)
            if (root.left is None and root.right) or (root.left and root.right is None):
                half_nodes += 1
            if root.left:
                queue.append(root.left)
            if root.right:
                queue.append(root.right)
        return half_nodes
        
    def identitical(self, root1, root2):
        if root1 is None and root2 is None:
            return True
        if root1 is not None and root2 is not None:
            return (root1.val == root2.val) and self.identitical(root1.left, root2.left) and self.identitical(root1.right, root2.right)
        return False

    def diameter_of_tree(self, root) :
        if root is None:
            return 0

        lheight = self.height_of_binary_tree(root.left)
        rheight = self.height_of_binary_tree(root.right)

        ldiameter = self.diameter_of_tree(root.left)
        rdiameter = self.diameter_of_tree(root.right)
    
        diameter = max(lheight+rheight+1, max(ldiameter, rdiameter))
    
        return diameter
        
    def max_level_sum(self, root):
        if root is None:
            return 0
        queue = []
        queue.append(root)
        queue.append('$')
        max_sum = -1
        level_sum = 0
        while queue:
            root = queue.pop(0)
            if root == '$':
                if queue:
                    if level_sum > max_sum:
                        max_sum = level_sum
                    level_sum = 0
                    queue.append('$')
            else:
                level_sum += root.val
                if root.left:
                    queue.append(root.left)
                if root.right:
                    queue.append(root.right)
        return max_sum
                
    def print_root_to_leaf_path(self, root, arr):
        if root is None:
            return 0
        arr.append(root.val)
        left = self.print_root_to_leaf_path(root.left, arr)
        right = self.print_root_to_leaf_path(root.right, arr)
        
        if left == 0 and right == 0:
            print(arr)
        arr.pop()
        return 1

    def has_path_sum(self, root, input_sum):
        if root is None:
            return False
        if root.val == input_sum:
            return True
        left = self.has_path_sum(root.left, input_sum-root.val)
        right = self.has_path_sum(root.right, input_sum-root.val)
        if left or right:
            return True
        return False

    def sum_of_all_nodes(self, root):
        if root is None:
            return 0
        return root.val + self.sum_of_all_nodes(root.left) + self.sum_of_all_nodes(root.right)

    def check_duplicate_in_tree(self, root, d):
        if root is None:
            return 0
        if d.get(root.val):
            return True
        else:
            d[root.val] = 1
        left = self.check_duplicate_in_tree(root.left, d)
        right = self.check_duplicate_in_tree(root.right, d)
        if left or right:
            return True
        return False
        
        

tt_obj = TreeQuestions()
result = tt_obj.find_max_element(root)
print(result)
result = tt_obj.search_element(root, 38)
print(result)
result = tt_obj.sizeoftree(root)
print(result)
result = tt_obj.level_order_reverse_order(root)
print(result)
result = tt_obj.height_of_binary_tree(root)
print(result)
result = tt_obj.deepest_node_binary_tree(root)
print(result)
result = tt_obj.deepest_node_binary_tree_optimised(root)
print(result)
result = tt_obj.number_of_leaves_recursion(root)
print(result)
result = tt_obj.number_of_leaves_iterative(root)
print(result)
result = tt_obj.number_of_full_nodes_recursion(root)
print(result)
result = tt_obj.number_of_full_nodes_iterative(root)
print(result)
#result = tt_obj.number_of_half_nodes_recursive(root)
#print(result)
result = tt_obj.number_of_half_nodes_iterative(root)
print(result)
result = tt_obj.identitical(root, root2)
print(result)
result = tt_obj.diameter_of_tree(root)
print(result)
result = tt_obj.max_level_sum(root)
print(result)
result = tt_obj.print_root_to_leaf_path(root, [])
print(result)
result = tt_obj.has_path_sum(root, 44)
print(result)
result = tt_obj.sum_of_all_nodes(root)
print(result)
result = tt_obj.check_duplicate_in_tree(root, {})
print(result)
```

```python
# LC - 144
# Given the root of a binary tree, return the Preorder traversal of its nodes' values.
# output : You need to return list of all the traversed nodes.
# [1 5 4 8 13 15 23 21 38 6 41 21 7]


def preorder_traverse(root, result):
    if root is None:
        return None 
    result.append(root.data)
    preorder_traverse(root.left, result)
    preorder_traverse(root.right, result)
    return result

def preorder(root):
    result = preorder_traverse(root,[])
    print(result)
    return result


# ----------------------------------------

def preorder_traversal_iterative(root):
    if root is None:
        return root
    stack = []
    result = []
    while 1:
        while root is not None :
            result.append(root.data)
            stack.append(root)
            root = root.left

        if not stack:
            break
        root = stack.pop()
        root = root.right
    del(stack)
    print(result)
    return result
```
